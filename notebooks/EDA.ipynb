```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Прямая ссылка на классический датасет Churn Modelling (с GitHub)
url = 'https://raw.githubusercontent.com/sharmaroshan/Churn-Modelling-Dataset/master/Churn_Modelling.csv'

# Скачиваем данные прямо из интернета
print("Скачивание датасета...")
df = pd.read_csv(url)

# Сразу сохраняем локально, чтобы использовать в следующих ноутбуках!
df.to_csv('Churn_Modelling.csv', index=False)
print("Датасет успешно скачан и сохранен как 'Churn_Modelling.csv'!")

# Выводим первые 5 строк, чтобы посмотреть на данные
display(df.head())
```

```python
# Смотрим общую информацию: типы данных и количество непустых значений
print("=== Информация о датасете ===")
df.info()

# Проверяем наличие пропусков (Missing values)
print("\n=== Количество пропусков в каждой колонке ===")
print(df.isna().sum())

# Вывод основных статистик по числовым колонкам (возраст, баланс и т.д.)
display(df.describe())
```

```python
# Настраиваем размер графика
plt.figure(figsize=(8, 5))

# Подсчитываем количество ушедших и оставшихся клиентов
churn_counts = df['Exited'].value_counts()

# Строим столбчатую диаграмму (Bar plot)
# 0 - клиент остался, 1 - клиент ушел (отток)
bars = plt.bar(['Остались (0)', 'Ушли (1)'], churn_counts.values, color=['#4CAF50', '#F44336'], edgecolor='black')

plt.title('Распределение целевой переменной', fontsize=14, pad=15)
plt.ylabel('Количество клиентов', fontsize=12)

# Добавляем текстовые подписи с количеством над каждым столбцом
for bar in bars:
    yval = bar.get_height()
    # Вычисляем процент
    percentage = f"{(yval / len(df) * 100):.1f}%"
    # Добавляем текст
    plt.text(bar.get_x() + bar.get_width()/2, yval + 100, 
             f"{int(yval)}\n({percentage})", 
             ha='center', va='bottom', fontsize=11, fontweight='bold')

# Увеличим верхний предел оси Y, чтобы текст не обрезался
plt.ylim(0, max(churn_counts.values) * 1.15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print("ВЫВОД:")
print("На графике четко виден дисбаланс классов: ~80% клиентов остались и ~20% ушли.")
print("Это означает, что при обучении моделей нам обязательно нужно будет использовать балансировку (например, параметр class_weight).")
```
